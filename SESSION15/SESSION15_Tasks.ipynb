```python
#TASK 1: Safe Division Function

def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return 'Cannot divide by zero'

print("10 divided by 2:", safe_divide(10, 2))
print("10 divided by 0:", safe_divide(10, 0))
```

```python
#TASK 2: Zomato Rating Input Validation

try:
    reviews_input = input("Enter number of reviews: ")
    stars_input = input("Enter total stars received: ")
    
    reviews = int(reviews_input)
    total_stars = int(stars_input)
    
    avg_rating = total_stars / reviews
    print("Average Rating is:", avg_rating)

except ValueError:
    print("Error: Please enter numbers only! Letters or symbols are not allowed.")
except ZeroDivisionError:
    print("Error: Reviews count cannot be zero.")
```

```python
#TASK 3: Custom Exception for Playlist

class InvalidDurationError(Exception):
    pass

def get_playlist_duration(songs):
    total_seconds = 0
    for duration in songs:
        if duration < 0:
            raise InvalidDurationError("Song duration cannot be negative!")
        total_seconds = total_seconds + duration
    
    total_minutes = total_seconds / 60
    return total_minutes

good_playlist = [180, 120, 240]
print("Total Minutes:", get_playlist_duration(good_playlist))

bad_playlist = [180, -90, 240]
try:
    print(get_playlist_duration(bad_playlist))
except InvalidDurationError as error:
    print("Caught custom error successfully:", error)
```

```python
#TASK 4: Flipkart Order with Else and Finally

try:
    price_input = input("Enter item price: ")
    qty_input = input("Enter item quantity: ")
    
    price = float(price_input)
    quantity = int(qty_input)
    
except ValueError:
    print("Error: Invalid input. Price must be a decimal and quantity must be a whole number.")

else:
    total_price = price * quantity
    print("Total Order Price: ₹", total_price)

finally:
    print("Thank you for shopping!")
```
