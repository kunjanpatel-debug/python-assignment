```python
#TASK 3: Adding the play_preview Method

class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play_preview(self):
        print("Playing 30-second preview of " + self.title + " by " + self.artist)

track = Song("Believer", "Imagine Dragons", 204)

track.play_preview()
```

```python
#TASK 4: FoodOrder Class with add_item Method

class FoodOrder:
    def __init__(self, restaurant_name):
        self.restaurant_name = restaurant_name
        self.items = []
        self.total_price = 0

    def add_item(self, item, price):
        self.items.append(item)
        self.total_price = self.total_price + price

my_order = FoodOrder("La Pino'z Pizza")
my_order.add_item("7-Cheese Pizza", 350)
my_order.add_item("Garlic Bread", 120)

print("Restaurant:", my_order.restaurant_name)
print("Items Ordered:", my_order.items)
print("Total Price: ₹", my_order.total_price)
```

```python
#TASK 5: Refactored Song Class with play_count

class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration
        self.play_count = 0  

    def increment_play_count(self):
        self.play_count = self.play_count + 1

my_track = Song("Shape of You", "Ed Sheeran", 233)

print("Initial play count:", my_track.play_count)

my_track.increment_play_count()
my_track.increment_play_count()
my_track.increment_play_count()

print("Updated play count:", my_track.play_count)
```
