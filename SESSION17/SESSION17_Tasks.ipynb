```python
#TASK 1: Math Module Calculations

import math

num = 16
fact_num = 5

print("Square root of", num, ":", math.sqrt(num))
print("Factorial of", fact_num, ":", math.factorial(fact_num))
print("Value of Pi:", math.pi)
```

```python
#TASK 2: Listing Image Files in Folder

import os

all_files = os.listdir('.')

image_files = []
for file_name in all_files:
    if file_name.lower().endswith('.jpg') or file_name.lower().endswith('.png'):
        image_files.append(file_name)

if image_files:
    print("Found Image Files:", image_files)
else:
    print("No .jpg or .png image files found in this directory.")
```

```python
#TASK 3: Day of Week from Date String

from datetime import datetime

date_input = input("Enter a date (YYYY-MM-DD): ")  

try:
    parsed_date = datetime.strptime(date_input, "%Y-%m-%d")
    
    day_name = parsed_date.strftime("%A")
    print(f"The day of the week for {date_input} is: {day_name}")

except ValueError:
    print("Error: Invalid date format! Please use YYYY-MM-DD format.")
```

```python
#TASK 4: Custom Module (insta_utils.py)

module_content = """
def format_follower_count(n):
    if n >= 1000000:
        return str(n // 1000000) + 'M'
    elif n >= 1000:
        return str(n // 1000) + 'K'
    else:
        return str(n)
"""

file = open("insta_utils.py", "w")
file.write(module_content.strip())
file.close()

import insta_utils

print("1500 followers  :", insta_utils.format_follower_count(1500))
print("2300000 followers :", insta_utils.format_follower_count(2300000))
print("850 followers   :", insta_utils.format_follower_count(850))
```
