```python
#TASK 1: Reverse a String Using Recursion

def reverse_string(s):
    if len(s) <= 1:
        return s
    return s[-1] + reverse_string(s[:-1])

print("Reversed 'hello':", reverse_string("hello"))
print("Reversed 'Zomato':", reverse_string("Zomato"))
```

```python
#TASK 2: Playlist Duration Total via Recursion

def playlist_durations(durations):
    if len(durations) == 0:
        return 0
    return durations[0] + playlist_durations(durations[1:])

song_times = [180, 210, 240, 150]  

print("Total Playlist Duration in seconds:", playlist_durations(song_times))
```

```python
#TASK 3: Scope Analysis (Local vs Global)

count = 10  

def update_count():
    count = 5  
    print('Inside:', count)

update_count()
print('Outside:', count)

#Explanation of what prints and why:
#1. 'Inside: 5' prints because the function looks at its own local variable 'count'.
#2. 'Outside: 10' prints because the local variable inside the function dies when the function ends,
#leaving the global variable 'count' completely unchanged.
```

```python
#TASK 4: Counting Instagram Likes

def count_likes(posts_list):
    total = 0
    
    for post in posts_list:
        total = total + post['likes']
        
        if 'replies' in post:
            total = total + count_likes(post['replies'])
            
    return total

my_instagram_data = [
    {
        'likes': 100,
        'replies': [
            {'likes': 20},
            {'likes': 10}
        ]
    }
]

print("Total Likes:", count_likes(my_instagram_data))
```

```python
# TASK 5: Global vs Local Variables

app_status = "WhatsApp is Open"  

def check_status():
    user_status = "Online" 
    print("Inside the function:")
    print("Global variable works here :", app_status)
    print("Local variable works here :", user_status)

print("Before calling the function:")
print("Global variable :", app_status)
print("-" * 30)

check_status()
print("-" * 30)

print("After calling the function:")
print("Global variable still works :", app_status)
print("Local variable : Cannot be accessed anymore (it is deleted!)")
```
