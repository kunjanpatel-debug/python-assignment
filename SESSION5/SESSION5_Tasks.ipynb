```python
# TASK 3: Zomato Order Pop Function

def remove_last_item(order_list):
    removed = order_list.pop()
    return removed

my_orders = ['Pizza', 'Burger', 'Fries', 'Coke']
item_removed = remove_last_item(my_orders)

print("Removed item:", item_removed)
print("Remaining orders in list:", my_orders)
```
