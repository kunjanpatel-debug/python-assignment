```python
#TASK 2: Reading a File and Printing Uppercase

file = open("my_playlist.txt", "r")

for line in file:
    song = line.strip()
    print(song.upper())

file.close()
```

```python
#TASK 3: Handling a CSV File for IPL Scores

import csv

csv_file = open("ipl_scores.csv", "w", newline="")
writer = csv.writer(csv_file)
writer.writerow(["Match", "Team1", "Team2", "Winner"])  
writer.writerow(["Match 1", "CSK", "RCB", "CSK"])
writer.writerow(["Match 2", "MI", "GT", "GT"])
writer.writerow(["Match 3", "KKR", "SRH", "KKR"])
csv_file.close()

read_file = open("ipl_scores.csv", "r")
reader = csv.reader(read_file)

next(reader)

for row in reader:
    match_name = row[0]
    winning_team = row[3]
    print(match_name, "Winner is:", winning_team)

read_file.close()
```
