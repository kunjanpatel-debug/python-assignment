```python
#TASK 1: Variable Declaration and Data Type Verification

followers_count = 1250
average_rating = 4.6
app_name = "Zomato"
is_premium_user = True

print("Followers:", followers_count)
print("Data Type:", type(followers_count))
print("Avg Rating:", average_rating, "| Data Type:", type(average_rating))
print("App Name:", app_name, "| Data Type:", type(app_name))
print("Premium Status:", is_premium_user, "| Data Type:", type(is_premium_user))
```

```python
#TASK 2: Zomato Bill Type Casting and GST Calculation

order_price = input("Enter Zomato order price: ")

order_price = float(order_price)

gst_amount = order_price * 0.18
final_bill = order_price + gst_amount

print(f"Base Order Value: {order_price:.2f}")
print(f"GST Amount (18%): {gst_amount:.2f}")
print(f"Total Payable Bill Amount: {final_bill:.2f}")
```

```python
#TASK 3: Flipkart E-Commerce Cart Value Aggregation

flipkart_prices = ['199.99', '299.50', '150']

flipkart_prices = [float(price) for price in flipkart_prices]

total_cart_value = sum(flipkart_prices)

print("Converted Floats List:", flipkart_prices)
print(f"Aggregated Flipkart Total Cart Value: {total_cart_value:.2f}")
```

```python
#TASK 4: Conditional Order Discount Function Thresholds

def is_discount_applicable(order_amount):
    return order_amount > 500

test_amount_1 = 450.0
test_amount_2 = 750.0

print(f"Is discount applicable for {test_amount_1}? -> {is_discount_applicable(test_amount_1)}")
print(f"Is discount applicable for {test_amount_2}? -> {is_discount_applicable(test_amount_2)}")
```

```python
# TASK 5: Spotify Review Dataset Type Casting and Extrema Filtering

spotify_ratings_str = ['4.5', '3.0', '5', '4.2']

spotify_ratings_float = []
for rating in spotify_ratings_str:
    spotify_ratings_float.append(float(rating))

highest_rating = max(spotify_ratings_float)

print("Processed Dataset Array:", spotify_ratings_float)
print("Peak Rating Value Extracted:", highest_rating)
```
