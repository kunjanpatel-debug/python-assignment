```python
#TASK 1: Product Name String Case Transformations

product = "Redmi Note 12 Pro"

uppercase_name = product.upper()
lowercase_name = product.lower()

print("Original String:", product)
print("Uppercase View: ", uppercase_name)
print("Lowercase View: ", lowercase_name)
```

```python
#TASK 2: Brand Name Text Normalization Function

def clean_brand_name(name):
    return name.strip().replace('-', ' ')

test_brand = "  oneplus-Nord  "
cleaned_result = clean_brand_name(test_brand)

print(f"Raw Input: '{test_brand}'")
print(f"Cleaned Output: '{cleaned_result}'")
```

```python
#TASK 3: E-Commerce Product Metadata Extraction via Slicing

full_product_title = "Apple iPhone 14 Pro Max"

words = full_product_title.split()
brand_word_length = len(words[0])

extracted_brand = full_product_title[:brand_word_length]
extracted_model = full_product_title[brand_word_length + 1:]

print("Source String:", full_product_title)
print("Extracted Brand Name:", extracted_brand)
print("Extracted Model Name:", extracted_model)
```

```python
#TASK 4: Standardized Product Card Display String Generation

def format_product_display(name, price):
    return f"{name} : {price}"

formatted_string = format_product_display("Boat Earbuds", 1299)

print("Generated Output Standard:", formatted_string)
```

```python
#TASK 5: Batch Processing and Sanitization of Messy Datasets

messy_products = ['   mi-Band 5   ', '  SAMSUNG-Galaxy  ', ' realme-Book ']
cleaned_products = []

for target_item in messy_products:
    sanitized_item = target_item.strip().replace('-', ' ').title()
    cleaned_products.append(sanitized_item)

print("Original Dirty Dataset List:", messy_products)
print("Sanitized Production Array: ", cleaned_products)
```
