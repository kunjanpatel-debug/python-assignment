```python
#TASK 1: BookMyShow Movie List Iterator

movies = ["Kalki", "Jawan", "Bang Bang", "Mission Impossible", "Animal"]

movie_iterator = iter(movies)

print(next(movie_iterator))
print(next(movie_iterator))
print(next(movie_iterator))
print(next(movie_iterator))
print(next(movie_iterator))
```

```python
#TASK 2: Tracklist Position with Enumerate

playlist = ["Timeless", "Believer", "No Loss", "Blinding Lights", "Excuses", "Hukum"]

for track_num, song in enumerate(playlist, start=1):
    print(f"{track_num}. {song}")
```

```python
#TASK 3: Combining Items and Prices

food_items = ["Pizza", "Burger", "Pasta", "Cold Coffee"]
prices = [250, 120, 180, 90]

for item, price in zip(food_items, prices):
    print(f"{item} - {price}")
```

```python
#TASK 4: Instagram Caption Generator

def insta_posts_generator(posts):
    for caption in posts:
        yield caption

captions_list = ["Weekend vibes!", "Tea first, always", "Chasing sunsets"]

caption_gen = insta_posts_generator(captions_list)

try:
    print(next(caption_gen))
    print(next(caption_gen))
    print(next(caption_gen))
    print(next(caption_gen))
except StopIteration:
    print("All post captions have been printed!")
```

```python
#TASK 5: Paytm Cashback Generator

def cashback_generator(transactions):
    for amount in transactions:
        cashback = amount * 0.05
        yield cashback

paytm_transactions = [100, 500, 1200, 2500]

for cb in cashback_generator(paytm_transactions):
    print(f"Cashback earned: ₹{cb:.2f}")
```
