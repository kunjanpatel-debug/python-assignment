```python
#TASK 1: Zomato Order Total Calculation

order_amounts = [120, 250, 90, 310, 150]

total_value = 0

for amount in order_amounts:
    total_value = total_value + amount

print("Total order value is: ", total_value)
```

```python
#TASK 2: Printing Cricket Scores Until Century

cricket_scores = [45, 78, 102, 34, 67, 89]
index = 0

while index < len(cricket_scores):
    score = cricket_scores[index]
    
    if score > 100:
        break
        
    print("Score:", score)
    index = index + 1
```

```python
# TASK 3: Flipkart Cart Filtering

item_prices = [299, 499, 199, 999, 149]
remaining_total = 0

for price in item_prices:
    if price < 200:
        continue
    
    remaining_total = remaining_total + price

print("Total of items priced 200 or above: ", remaining_total)
```
