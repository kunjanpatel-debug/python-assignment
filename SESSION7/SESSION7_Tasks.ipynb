```python
#TASK 1: IPL Ticket Eligibility Check

age_input = input("Enter your age: ")
age = int(age_input)

if age >= 18:
    print("Eligible for IPL ticket booking")
else:
    print("Not eligible")
```

```python
#TASK 2: Social Media Influencer Tier Check

followers_input = input("Enter your follower count: ")
followers = int(followers_input)

if followers < 10000:
    print("Micro Influencer")
elif followers >= 10000 and followers <= 100000:
    print("Rising Star")
else:
    print("Celebrity")
```

```python
#TASK 3: Zomato Delivery Fee Check

order_total_input = input("Enter your Zomato order total: ")
order_total = float(order_total_input)

if order_total > 299:
    print("Apply Free Delivery")
elif order_total >= 200 and order_total <= 299:
    print("Add more items for free delivery")
else:
    print("Delivery charges apply")
```

```python
#TASK 4: Flipkart Cashback Checker

cart_value_input = input("Enter Flipkart cart value: ")
cart_value = float(cart_value_input)

payment_method = input("Enter payment method (UPI, Card, Cash): ")

if cart_value > 1000:
    if payment_method == "UPI":
        print("Eligible for 10% cashback")
    else:
        print("Eligible for 5% cashback")
else:
    print("No cashback")
```
