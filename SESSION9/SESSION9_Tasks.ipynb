```python
#TASK 1: Basic Discount Calculation Function

def calculate_final_price(price, discount_rate):
    discount_amount = price * discount_rate
    final_price = price - discount_amount
    return final_price

result_1 = calculate_final_price(1200, 0.15)

print("Final Price after discount: Rs.", result_1)
```

```python
#TASK 2: Delivery Charge with Default City

def get_delivery_charge(amount, city='Ahmedabad'):
    if city == 'Ahmedabad':
        return 30
    else:
        return 50

charge_default = get_delivery_charge(300)
print("Delivery charge for Ahmedabad: Rs.", charge_default)

charge_custom = get_delivery_charge(300, city='Vadodara')
print("Delivery charge for Vadodara: Rs.", charge_custom)
```

```python
#TASK 3: Coupon Message Formatting

def format_coupon_message(username, discount=10):
    message = "Hi " + username + ", you get " + str(discount) + "% off!"
    return message

msg_default = format_coupon_message("Kunjan")
print(msg_default)

msg_custom = format_coupon_message("Dheer", discount=20)
print(msg_custom)
```

```python
#TASK 4: Apply Default Discount Rate

def apply_discount(price, rate=0.10):
    discount_amount = price * rate
    final_price = price - discount_amount
    return final_price

discounted_price = apply_discount(1500)
print("Final price using default 10% rate: Rs.", discounted_price)
```

```python
# TASK 5: Cashback Calculation Function

def calculate_cashback(amount, cashback_rate=0.05):
    cashback_amount = amount * cashback_rate
    return cashback_amount

zomato_cashback = calculate_cashback(500)
print("Zomato Cashback (Default 5%): Rs.", zomato_cashback)

flipkart_cashback = calculate_cashback(2000, cashback_rate=0.07)
print("Flipkart Cashback (Custom 7%): Rs.", flipkart_cashback)
```
