```python
#TASK 1: Creating a Playlist Price Dictionary

playlist_prices = {
    "Lo-Fi Beats": 199,
    "Bollywood Hits": 299,
    "Gym Motivation": 249,
    "Chill Vibes": 150,
    "Classic Rock": 399
}

print("Initial playlist prices:", playlist_prices)
```

```python
#TASK 2: Updating a Playlist Price

def update_playlist_price(playlist, new_price):
    playlist_prices[playlist] = new_price

update_playlist_price("Chill Vibes", 180)

print("Updated playlist prices:", playlist_prices)
```

```python
# TASK 3: Deleting a Playlist

del playlist_prices["Classic Rock"]

print("Dictionary after deleting Classic Rock:", playlist_prices)
```

```python
#TASK 4: Zomato and Swiggy Restaurant Sets

zomato_restaurants = {"Burger King", "McDonalds", "Dominos", "Subway"}
swiggy_restaurants = {"Dominos", "Subway", "Pizza Hut", "KFC"}

all_restaurants = zomato_restaurants.union(swiggy_restaurants)

common_restaurants = zomato_restaurants.intersection(swiggy_restaurants)

print("Union:", all_restaurants)
print("Intersection :", common_restaurants)
```
