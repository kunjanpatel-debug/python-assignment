```python
#TASK 1: Product Class with Private Price

class Product:
    def __init__(self, name, price):
        self.name = name
        self._price = price  

    def display_price(self):
        print("Product:", self.name, "| Price: ", self._price)

item = Product("Boat Headphones", 1499)
item.display_price()
```

```python
#TASK 2: Getter and Setter with Validation

class Product:
    def __init__(self, name, price):
        self.name = name
        self._price = price

    def get_price(self):
        return self._price

    def set_price(self, new_price):
        if new_price < 0:
            raise ValueError("Price cannot be negative!")
        self._price = new_price

item = Product("Wireless Mouse", 500)

item.set_price(650)
print("Updated Price via Getter: ", item.get_price())

try:
    item.set_price(-200)
except ValueError as error:
    print("Error caught successfully:", error)
```

```python
#TASK 3: Encapsulating a Songs List

class Playlist:
    def __init__(self, name):
        self.name = name
        self._songs = []  

    def add_song(self, song):
        self._songs.append(song)

    def remove_song(self, song):
        if song in self._songs:
            self._songs.remove(song)

    def get_songs(self):
        return self._songs

my_playlist = Playlist("Chill Vibes")

my_playlist.add_song("Kesariya")
my_playlist.add_song("Believer")
my_playlist.add_song("Shape of You")
print("Current Playlist:", my_playlist.get_songs())

my_playlist.remove_song("Believer")
print("Playlist after removing 'Believer':", my_playlist.get_songs())
```

```python
# TASK 4: Abstract Class for Payment Methods

from abc import ABC, abstractmethod

class PaymentMethod(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class UPI(PaymentMethod):
    def pay(self, amount):
        print("Paid ₹" + str(amount) + " successfully using UPI scan.")

class CreditCard(PaymentMethod):
    def pay(self, amount):
        print("Paid ₹" + str(amount) + " successfully using Credit Card OTP.")

upi_payment = UPI()
upi_payment.pay(499)

card_payment = CreditCard()
card_payment.pay(1200)
```
