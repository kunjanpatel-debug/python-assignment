```python
#TASK 1: Lambda Function for 18% GST

add_gst = lambda price: price + (price * 0.18)

print("Price 100 with GST: Rs.", add_gst(100))
print("Price 250 with GST: Rs.", add_gst(250))
print("Price 500 with GST: Rs.", add_gst(500))
```

```python
#TASK 2: Cleaning Song Titles with map()

Unformated_songs = [' shape OF you ', '  KESARIYA ', 'blinding LIGHTS   ']

clean_songs = list(map(lambda title: title.strip().title(), Unformated_songs))

print("Cleaned Songs List:", clean_songs)
```

```python
#TASK 3: Filtering 'S' Products with filter()

products = ['Samsung Galaxy', 'iPhone 14', 'Sony Headphones', 'Shoes', 'MacBook']

s_products = list(filter(lambda name: name.lower().startswith('s'), products))

print("Products starting with S:", s_products)
```

```python
# TASK 4: Total Bill Using reduce()

from functools import reduce

order_amounts = [120, 340, 560, 80]

total_bill = reduce(lambda x, y: x + y, order_amounts)

print("Total Zomato Bill Amount: Rs.", total_bill)
```

```python
# TASK 5: Combined Map, Filter, and Reduce

from functools import reduce

numbers_list = [40, 60, 80, 120]

doubled_numbers = map(lambda x: x * 2, numbers_list)

filtered_numbers = filter(lambda x: x > 100, doubled_numbers)

final_sum = reduce(lambda x, y: x + y, filtered_numbers)

print("Final processed sum result:", final_sum)
```
